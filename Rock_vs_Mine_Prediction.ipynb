# A Machine Learning project to predict if an object is a Rock or a mine using Sonar senor data.

This is one of my projects for ML learning.

## Importing the libraries and dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Data Collections and Preprocessing

```python
#Read the file and save it as a Pandas dataframe
sonar_data_df = pd.read_csv('/Sonar data.csv', header=None)
sonar_data_df.head()
```

```python
# Describe the data
sonar_data_df.describe()
```

```python
# get the value counts of the categorical data at the 60th column
sonar_data_df[60].value_counts()
```

```python
# Group by Rock or Mine
sonar_data_df.groupby(60).mean()
```

```python
# Separating the data and label
X = sonar_data_df.drop(columns=60, axis=1)
y = sonar_data_df[60]
```

```python
# lets print the Features in X
print(X)
```

```python
# Lets Print he labels in Y
print(y)
#print([R for R in Y ])
```

## Training and Test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, stratify=y, random_state=42)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Training a Machine Learning Model

```python
# We use Logistic Regression Model

model = LogisticRegression()
```

# Fit the model

```python
model.fit(X_train, y_train)
```

Evaluate our model

```python
#Accuracy of training data

X_train_prediction = model.predict(X_train)

training_data_accuracy = accuracy_score(X_train_prediction, y_train)
```

```python
print('Accuracy on training data: ', training_data_accuracy)
```

Evaluate our model on the data it has not 'seen'

```python
#Accuracy on test data

X_test_prediction = model.predict(X_test)

test_data_accuracy = accuracy_score(X_test_prediction, y_test)
```

```python
print('Accuracy on testing data: ', test_data_accuracy)
```
